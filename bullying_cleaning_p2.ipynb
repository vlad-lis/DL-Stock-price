```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bullying_clean1.csv')
```

```python
df.rename(columns = {'Bullied_on_school_property_in_past_12_months': 'bullied_in_school',
                    'Bullied_not_on_school_property_in_past_12_months': 'bullied_outside_school',
                    'Cyber_bullied_in_past_12_months': 'cyberbullied',
                    'Custom_Age': 'age'}, inplace=True)

df.drop(['record',
         'Missed_classes_or_school_without_permission',
        'Most_of_the_time_or_always_felt_lonely'], axis=1, inplace=True)

df.columns = [c.lower() for c in df.columns]
```

```python
df['age'].unique()
```

```python
def clean_age(value):
    if value == '11 years old or younger':
        return 11
    elif value == '12 years old':
        return 12
    elif value == '13 years old':
        return 13
    elif value == '14 years old':
        return 14
    elif value == '15 years old':
        return 15
    elif value == '16 years old':
        return 16
    elif value == '17 years old':
        return 17
    elif value == '18 years old or older':
        return 18
    else:
        print('something went wrong')
        return

df['age'] = df['age'].apply(lambda x: clean_age(x))
```

```python
def clean_binary(value):
    if value == 'Yes':
        return 1
    elif value == 'No':
        return 0
    else:
        print('something went wrong')
        return

#target variables
df['bullied_in_school'] = df['bullied_in_school'].apply(lambda x: clean_binary(x))
df['bullied_outside_school'] = df['bullied_outside_school'].apply(lambda x: clean_binary(x))
df['cyberbullied'] = df['cyberbullied'].apply(lambda x: clean_binary(x))

#other binary features
df['were_underweight'] = df['were_underweight'].apply(lambda x: clean_binary(x))
df['were_overweight'] = df['were_overweight'].apply(lambda x: clean_binary(x))
df['were_obese'] = df['were_obese'].apply(lambda x: clean_binary(x))
```

```python
def clean_attacked(value):
    if value == '0 times':
        return 0
    elif value == '1 time':
        return 1
    elif value == '2 or 3 times':
        return 2
    elif value == '4 or 5 times':
        return 3
    elif value == '6 or 7 times':
        return 4
    elif value == '8 or 9 times':
        return 5
    elif value == '10 or 11 times':
        return 6
    elif value == '12 or more times':
        return 7
    else:
        print('something went wrong')
        return
    
df['physically_attacked'] = df['physically_attacked'].apply(lambda x: clean_attacked(x))
```

```python
def clean_fighting(value):
    if value == '0 times':
        return 0
    elif value == '1 time':
        return 1
    elif value == '2 or 3 times':
        return 2
    elif value == '4 or 5 times':
        return 3
    elif value == '6 or 7 times':
        return 4
    elif value == '8 or 9 times':
        return 5
    elif value == '10 or 11 times':
        return 6
    elif value == '12 or more times':
        return 7
    else:
        print('something went wrong')
        return
    
df['physical_fighting'] = df['physical_fighting'].apply(lambda x: clean_fighting(x))
    
```

```python
def clean_sex(value):
    if value == 'Male':
        return 0
    elif value == 'Female':
        return 1
    else:
        print('something went wrong')
        return

df['sex'] = df['sex'].apply(lambda x: clean_sex(x))
```

```python
def clean_lonely(value):
    if value == 'Never':
        return 0
    elif value == 'Rarely':
        return 1
    elif value == 'Sometimes':
        return 2
    elif value == 'Most of the time':
        return 3
    elif value == 'Always':
        return 4
    else:
        print('something went wrong')
        return
    
df['felt_lonely'] = df['felt_lonely'].apply(lambda x: clean_lonely(x))
    
```

```python
def clean_friends(value):
    if value == '0':
        return 0
    elif value == '1':
        return 1
    elif value == '2':
        return 2
    elif value == '3 or more':
        return 3
    else:
        print('something went wrong')
        return
    
df['close_friends'] = df['close_friends'].apply(lambda x: clean_friends(x))
```

```python
def clean_attendance(value):
    if value == '0 days':
        return 0
    elif value == '1 or 2 days':
        return 1
    elif value == '3 to 5 days':
        return 2
    elif value == '6 to 9 days':
        return 3
    elif value == '10 or more days':
        return 4
    else:
        print('something went wrong')
        return
    
df['miss_school_no_permission'] = df['miss_school_no_permission'].apply(lambda x: clean_attendance(x))
    
```

```python
def clean_compassion(value):
    if value == 'Never':
        return 0
    elif value == 'Rarely':
        return 1
    elif value == 'Sometimes':
        return 2
    elif value == 'Most of the time':
        return 3
    elif value == 'Always':
        return 4
    else:
        print('something went wrong')
        return
    
df['other_students_kind_and_helpful'] = df['other_students_kind_and_helpful'].apply(lambda x: clean_compassion(x))
df['parents_understand_problems'] = df['parents_understand_problems'].apply(lambda x: clean_compassion(x))
    
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
```

```python
df.to_csv('bullying_clean2.csv', index=False)
```
