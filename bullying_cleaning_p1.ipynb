```python
import pandas as pd
import numpy as np
```

```python
raw_df = pd.read_csv('bullying_2018.csv')
```

```python
raw_df['all values'] = raw_df
raw_df['list of values'] = raw_df['all values'].apply(lambda x: x.split(';'))
raw_df['len'] = raw_df['list of values'].apply(len)
len(raw_df['all values'])
```

```python
columns_list = list(raw_df.columns[0].split(';'))
raw_df.drop(raw_df.columns[0], axis=1, inplace=True)
```

```python
df = pd.DataFrame(index=range(56981), columns=columns_list)
```

```python
for column in columns_list:
    col_index = columns_list.index(column)
    
    for i in range(0, 56981):
        df.iloc[i, col_index] = raw_df.iloc[i, 1][col_index]
```

```python
df.replace(' ', np.nan, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv('bullying_clean1.csv', index=False)
```
